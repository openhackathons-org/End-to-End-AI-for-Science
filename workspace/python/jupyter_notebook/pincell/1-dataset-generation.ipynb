# 1. Dataset generation

In this notebook, we generate the dataset used for model training. 

```python
import os

import pandas as pd
import yaml

from src.pincell import Pincell
from src.simulation import Simulation
from src.utils import generate_pincells
```

## Cross section data

Download the [`ENDF/B-VII.1`](https://openmc.org/data/) cross section data library to your local machine and configure the `openmc_cross_sections_path` variable in [`config.yaml`](config.yaml)

The data library contains neutron interaction cross sections, thermal scattering, photoatomic, and atomic relaxation data across multiple temperatures, which OpenMC uses to model particle transport and material behaviour accurately.

```python
with open("config.yaml") as f:
    config = yaml.safe_load(f)

os.environ["OPENMC_CROSS_SECTIONS"] = config["openmc_cross_sections_path"]
```

## Design of Experiments

Generate a dataset that provides broad coverage of the design space by varying the following design parameters:

- Fuel enrichment
- Geometry:
  - $p$ - pin pitch (lattice pitch)
  - $r_f$ - fuel rod radius
  - $r_g$ - gap radius
  - $r_c$ - cladding thickness
  <!-- - $r_{cell}$ - pin cell radius (coolant radius) -->

<!-- ![](figures/pin-cell.png) -->

The function `generate_pincells()` employs [latin hypercube sampling](https://en.wikipedia.org/wiki/Latin_hypercube_sampling) to produce a broadly distributed set of input parameter combinations. Users can modify the number of samples and the parameter bounds.

```python
n_samples = 200
bounds = [
    (2.0, 5.0),     # enrichment
    (0.1, 1.0),     # fuel radius
    (0.001, 0.02),  # gap thickness
    (0.05, 0.1),    # cladding thickness
]
pitch_max = 4.0

enrichments, pitches, fuel_radii, gap_thicknesses, clad_thicknesses = generate_pincells(
    n_samples, bounds, pitch_max, seed=42
)
```

Instantiate a list of `Pincell` objects, each representing a unique pin-cell. The corresponding input features for [surrogate model training](3-fourier-neural-operator.ipynb) are also saved in `data/input/npy`.

```python
pincells = [
    Pincell(
        instance_id=i + 1,
        enrichment=enrichments[i],
        pitch=pitches[i],
        fuel_radius=fuel_radii[i],
        gap_thickness=gap_thicknesses[i],
        clad_thickness=clad_thicknesses[i],
        save_geometry_figure=True,
    )
    for i in range(n_samples)
]
```

## Run simulations

Iterate over each `Pincell` and run a Monte Carlo particle transport simulation to compute the neutron flux field. The `Simulation` class abstracts the underlying `OpenMC` implementation, allowing users to modify the number of particles and the mesh resolution without interacting directly with the lower-level simulation setup.

The corresponding outputs are stored in the `data/output` directory, while summary metadata for each case are written to `data/metadata.csv`.

```python
simulation = Simulation(particles=100000, mesh_dim=(100, 100, 2))

output_path = "data/metadata.csv"

for i, pincell in enumerate(pincells):
    result = simulation.run(pincell)
    df_result = pd.DataFrame([result])

    df_result.to_csv(
        output_path,
        mode="a",
        header=not os.path.exists(output_path),
        index=False,
    )
```
