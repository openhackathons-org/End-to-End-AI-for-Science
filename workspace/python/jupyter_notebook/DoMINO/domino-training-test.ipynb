## **Part 1:** The DoMINO Architecture

Machine learning (ML) models have been proposed as surrogate models to speed up simulations, but they face limitations in terms of accuracy, scalability, and generalization to new geometries. DoMINO, a new ML model designed to address these challenges. DoMINO is a multi-scale, iterative neural operator that uses local geometric information to predict flow fields in large-scale simulations. It is specifically validated for the automotive aerodynamics use case, showcasing its scalability, accuracy, and ability to generalize across different simulation scenarios. 

The DoMINO model evaluates solution fields within a computational domain by leveraging geometry representations in STL file format. Let's walk through the DoMINO architecture step by step through the following key stages:
- **Global Geometry Representation:** This component learns the overall shape of the object. It takes the unstructured point cloud from an STL file and converts it into a structured, grid-like representation using multi-scale point convolutional kernels. These work like smart filters that operate at different spatial resolutions, capturing both fine details and the broad structure of the object. The network is iterative, meaning it can propagate surface information out into the surrounding 3D space—allowing the geometry's influence to extend well beyond the surface.
- **Local Geometry Representation:** Once the global shape is encoded, this component zooms into small subregions (subdomains) to extract localized geometry features. These give the model a clear understanding of the shape and physics in the immediate neighborhood of a prediction point. This local information is used by the Branch Net of the Aggregation NN (inspired by DeepONet), conditioning predictions on the exact spatial and geometric context around each sample point.
- **Aggregation Network:** This is the core prediction module, structured like a DeepONet with:
  - A Branch Net that processes local geometry and context (from the previous step).
  -  A Trunk Net that learns nonlinear basis functions, which represent how to combine the input features from neighboring points.


In DoMINO, instead of relying on a fixed grid like in traditional CFD, the model samples a target point i from the mesh or point cloud and dynamically builds a stencil by selecting nearby points j. These neighbors are typically chosen using spatial proximity methods like k-nearest neighbors (k-NN), radius-based search, or by perturbing the coordinates of i. This stencil defines which nearby points contribute to the prediction at i.

The key innovation lies in how the influence—or weight—of each neighboring point is determined. Unlike traditional finite volume methods, where stencil weights are fixed and manually designed, DoMINO uses a neural network called the Trunk Net to learn these weights dynamically. The Trunk Net generates nonlinear basis functions that are conditioned on local geometry, physical parameters, and spatial context. This makes the weights adaptive—changing based on the shape or flow conditions at each point. As a result, DoMINO can make context-aware predictions that are significantly more flexible and accurate than conventional methods.

Over the next few sections, we will look into this process in more detail.

## **Global Geometry Representation:**
The **Global Geometry Representation** refers to the overall shape and structure of the entire object or domain that you are modeling. This representation captures all the geometric details across the entire computational domain.

Here is a step-by-step explanation of this process:

**Step 1: Construct Bounding Boxes**
First, two distinct bounding boxes are defined to create structured domains for analysis.

* **Surface Bounding Box:** A tight-fitting box is created directly around the 3D geometry. This focused domain is used for high-fidelity analysis right at the object's surface.
* **Computational Bounding Box:** A much larger box is defined that encompasses the entire simulation domain (e.g., the virtual wind tunnel).

Both of these boxes are defined in the `conf.yaml` file and will be populated with structured grids.

* **Code Connection:** In the main `DoMINO.forward` method, these grids are passed in through the `data_dict`. The computational grid is `data_dict["grid"]` (aliased as `p_grid`), and the surface grid is `data_dict["surf_grid"]` (aliased as `s_grid`). These are used as inputs for the `self.geo_rep_volume` and `self.geo_rep_surface` modules, respectively.

---


**Step 2: Create Multi-Scale Features using Learnable Kernels**
Next, the geometric information from the unstructured point cloud is projected onto a structured grid using **learnable point convolution kernels**. This is the core of the multi-scale approach.

* **Differentiable Ball Queries:** The method uses a "ball" (a sphere) around various points to find neighbors. This is a "differentiable" operation, meaning the network can learn the optimal way to gather local geometric information during training.
* **Multi-Radius for Multi-Scale:** The key to capturing features at different scales is defining a list of radii in the `conf.yaml` file (e.g., `volume_radii: [0.1, 0.5]`). For **each radius** in the list, the model learns a completely separate set of point convolutional kernels.
    * **Small radii** learn to detect fine-grained details (sharp edges, small curves).
    * **Large radii** learn to detect coarse, general features (overall curvature, large flat areas).

* **Code Connection:** This is implemented within the `GeometryRep` class. Its `__init__` method creates a `nn.ModuleList` of `BQWarp` and `GeoConvOut` modules, one for each specified radius. The `forward` method then loops through these radii (`for j in range(len(self.radii)):`), using `self.bq_warp[j]` to find neighboring points and `self.geo_conv_out[j]` to process them into an initial feature grid `x_encoding_inter`.

This process creates multiple "channels" of geometric information, each corresponding to a different scale, representing both short- and long-range geometric interactions.

**Step 3: Define a Single, High-Resolution Grid**
The multi-scale features generated in the previous step are projected onto a **single, fixed-resolution grid** within each bounding box.

* **Misconception Alert:** The model does **not** use multiple grids of different resolutions. It uses one grid resolution for the surface bounding box and one for the computational box (which are typically set to be the same).
* **Grid Configuration:** A parameter like `interp_res: [256, 96, 64]` in the configuration file defines the dimensions `(nx, ny, nz)` of this single grid. It represents the number of points along the x, y, and z axes, not a list of different resolutions.

* **Code Connection:** The `interp_res` parameter from the config is read into the `DoMINO` model and stored as `self.grid_resolution`. This value is then passed to the `BQWarp` and `GeoConvOut` constructors inside `GeometryRep`, ensuring all modules operate on the same grid dimensions.

---

**Step 4: Process and Propagate Features with CNNs**

Once the multi-scale geometric features have been placed onto the structured grid, the next stage is to process them with a 3D Convolutional Neural Network (CNN). This is the true **propagation** step.

1.  **Initial Features:** The process starts with the grid that already contains the features projected by the multi-scale point convolution kernels. At this point, the geometric information is still very localized.
2.  **Iterative CNN Processing:** The `GeoProcessor` module, which is a UNet-like 3D CNN, takes this feature grid as input and iteratively applies its layers.
3.  **Spreading the Information:** This iterative refinement is what effectively **propagates** the geometric information across the grid. Think of it like ripples spreading in a pond. Each pass through the `GeoProcessor` allows the influence of the surface geometry to spread further into the computational domain, ensuring that grid points far away from the object are still aware of its shape.

* **Code Connection:** This is implemented by the iterative loop inside `GeometryRep.forward`: `for _ in range(self.hops):`. Inside the loop, the line `dx = self.geo_processors[j](x_encoding_inter) / self.hops` shows the application of the `GeoProcessor` CNN. The result is added back (`x_encoding_inter = x_encoding_inter + dx`), refining the features with each "hop".

**Step 5: Calculate Signed Distance Function (SDF) and its Gradients**

In parallel, a **Signed Distance Function (SDF)** and its gradients are calculated on the grid.

* The SDF provides crucial topological information: for any point on the grid, it tells you the shortest distance to the object's surface and whether the point is inside (negative), outside (positive), or on the surface (zero).
* These SDF values and their gradients are added as additional feature channels to the grid, enriching the network's understanding of the geometry.

* **Code Connection:** Inside `GeometryRep.forward`, the block `if self.geo_encoding_type == "both" or self.geo_encoding_type == "sdf":` handles this. It computes gradients using `torch.gradient(sdf, dim=[2, 3, 4])`, concatenates them with the original and scaled SDFs (`torch.cat(...)`), and processes them through a dedicated CNN, `self.geo_processor_sdf`, to create the `sdf_encoding`.

---

**Step 6: Combine Features for the Final Representation**

The final Global Geometry Representation is formed by combining all the calculated features into a single, comprehensive tensor. This includes:

* The processed multi-scale features from the different radius kernels (from Step 4).
* The Signed Distance Function and its gradient features (from Step 5).
* **Code Connection:** At the end of `GeometryRep.forward`, the line `encoding_g = torch.cat((x_encoding, sdf_encoding), 1)` performs this final combination. It concatenates the tensor from the point convolution stream (`x_encoding`) and the tensor from the SDF stream (`sdf_encoding`) along the channel dimension, creating the final multi-channel grid that represents the global geometry.

This final, rich representation contains all the necessary information about the object's shape, from fine details to coarse structure and topology, preparing it for the next stage of local analysis and final prediction.

## **Local Geometry Representation**

While the Global Geometry Representation provides the overall picture of the object, the **Local Geometry Representation** zooms in to understand the geometry in the **immediate vicinity** of a specific point `p` being evaluated in the simulation. This provides a detailed, high-fidelity context that is crucial for making accurate predictions.

The process for a single sampled point `p` is as follows:
1.  A **computational stencil** is formed by sampling neighboring points around `p`.
2.  A local "subregion" is defined around this stencil using a radius specified in the `conf.yaml` file (e.g., `geometry_local.volume_radii: [0.05, 0.1]`).
3.  The model then **extracts features from the pre-computed Global Geometry grid** that fall within this local subregion. This gives the point `p` a rich understanding of the specific geometric features right next to it.

* **Code Connection:** This entire process is encapsulated within the `DoMINO.geo_encoding_local` method. It takes the global feature grid `encoding_g` and the simulation mesh points `volume_mesh_centers` as input and produces the final local encoding that is fed into the Aggregation Network.

---

### BQWarp: Global vs. Local Representation

The `BQWarp` module is used in both stages, but its **purpose and direction are fundamentally different**. This is a key distinction. 

* **In Global Representation (Projection):** `BQWarp` maps from the **structured grid to the raw point cloud**. Its question is: "For each point on my organized grid, which unstructured STL vertices are nearby?" This projects the raw geometry *onto* the grid to create the global features in the first place.

* **In Local Representation (Sampling/Extraction):** `BQWarp` maps from the **unstructured simulation mesh to the structured global feature grid**. Its question is: "For my specific sampled point `p`, which features on the pre-computed global grid are within my local radius?" It samples or extracts from the global representation to create the local context.

* **Code Connection:** The difference in direction is explicitly controlled by the `reverse_mapping` flag passed to the `BQWarp.forward` method.
    * The **global** `BQWarp` (inside `GeometryRep`) defaults to `reverse_mapping=True`, finding grid points near the STL points.
    * The **local** `BQWarp` (called within `geo_encoding_local`) explicitly uses `reverse_mapping=False` to find features on the grid `p_grid` that are near the query points `volume_mesh_centers`.

This local sampling mechanism is set up in the `DoMINO` model's constructor:
```python
# BQ for surface
self.surface_neighbors_in_radius = (

        total_neighbors_in_radius = self.surface_neighbors_in_radius[ct] * (

**How Multi-Scale Global Features Affect Local Representation**

The interaction between the global and local stages is a key part of the DoMINO architecture. The global grid doesn't have different resolutions, but instead contains different **feature channels**, where each channel was created by a global kernel of a different radius (e.g., one channel for fine details from a 0.1 radius kernel, another for coarse features from a 0.5 radius kernel).

When the local representation module samples the global grid around a point `p`, it samples from **all of these feature channels simultaneously**.
This point `p` is an unstructured query point. It's a specific location in your simulation domain where you want to find the answer (e.g., "What is the velocity at this exact coordinate?"). These points typically come from the centers of the cells in your simulation mesh.

**The Detailed Explanation with an Analogy**
  
Think of the Global Geometry Grid as a highly detailed digital map of a city that you have already created and stored. The map has a fixed grid of pixels, and at each pixel, you have stored lots of information (elevation, population density, zoning type, etc.).
Point `p` is your current GPS location. Your location is a specific, arbitrary coordinate (latitude, longitude) and it almost never falls exactly on the center of one of the map's pixels.

In short:

1. Your phone (the "local module") takes your GPS coordinate (point `p`).
2. It looks at the digital map (the "global grid").
3. It finds the map pixels that are immediately around your location.
4. It samples the information from those nearby pixels (often using interpolation to get a precise value for your exact spot) to tell you what's at your location.

So, point `p` is the unstructured query point from the simulation mesh, and it is used to sample data from the structured global grid.

Now let see How Multi-Scale Global Features Affect Local Representation.
Think of the global grid as a multi-layered satellite map of a landscape.

* One layer is a high-resolution black-and-white photo, showing every rock and bush (**fine-detail channel**).
* Another layer is a color-coded elevation map, showing the general mountains and valleys (**coarse-feature channel**).

When you zoom in on a single GPS coordinate (the **local point `p`**), you get both the detailed photo and the general elevation for that exact spot.
When the model takes point p and samples the global grid, it doesn't just get one value. It gets a whole vector of features because it samples from all the feature channels simultaneously.

So, for that single point p, the model gets:

1. The interpolated value from the "fine-details" channel (learned with the small radius kernel).
2. The interpolated value from the "coarse-features" channel (learned with the large radius kernel).
3. The interpolated value from the "SDF" channel.
4. The interpolated values from the SDF gradient channels.

And so on...

This gives the model a rich, multi-scale description of the geometry in the immediate vicinity of point `p`, which is exactly what it needs to make a highly accurate, localized prediction.


* **Code Connection:** The `geo_encoding_local` method implements this by explicitly looping through the channel dimension of the global grid `encoding_g` (i.e., `for j in range(encoding_g.shape[1]):`). Inside the loop, it uses `torch.index_select` to sample the features from each specific channel based on the local neighborhood mapping. These sampled features from all channels are then concatenated (`torch.cat(encoding_g_inner, dim=2)`) before being processed by a final `LocalPointConv` layer. This is how the local representation inherits the multi-scale context. 

Therefore, the local representation knows not just that a surface is nearby, but also whether that surface is a finely detailed edge that is part of a larger, gently curving panel—all from a single sampling operation.

## **Aggregation and Solution Prediction**

After creating the global and local geometry representations, the final stage of the DoMINO architecture is to predict the physical solution (e.g., pressure, velocity). This is handled by an **Aggregation Network**, which uses a powerful structure inspired by the **Deep Operator Network (DeepONet)**.

A DeepONet works by splitting the problem into two parts, which are handled by two separate neural networks: a **Branch Net** and a **Trunk Net**.

---
### The Branch Net: Encoding Local Geometry

The job of the Branch Net is to understand the specific geometric context where a prediction needs to be made.

* **Input:** It takes the **local geometry encoding** as input. This is the rich, multi-scale feature set that was extracted from the global grid in the previous stage, describing the geometry in the immediate vicinity of a sampled point.

* **Purpose:** The Branch Net processes this information to create a single, powerful feature vector. This vector represents the distinct "shape" of the local neighborhood, conditioning the final prediction on the surrounding geometry.

* **Code Connection:** In the `DoMINO.forward` method, this local geometry encoding is computed by the `self.geo_encoding_local(...)` function. The result is stored in variables like `encoding_g_vol` or `encoding_g_surf`. This tensor is the final output of the Branch Net's processing chain and is later passed to the final aggregation step inside `calculate_solution`.

---
### The Trunk Net: Learning the Basis Functions

The job of the Trunk Net is to interpret the query itself. It learns non-linear "basis functions" that represent how the solution should behave at different locations in space.

* **Dynamic Stencil Creation:** To make a stable prediction at a central point `i`, the model dynamically creates a computational stencil.
    * For **volume predictions**, the `calculate_solution` function implements this by adding random noise to the coordinates of the central point `i` to sample neighboring points `j`. This is visible in the code with lines like `one_loop_bulk_random_noise = torch.rand_like(...)` and `one_loop_volume_m_c = volume_mesh_centers + one_loop_bulk_random_noise`.
    * For **surface predictions**, the `calculate_solution_with_neighbors` function often uses pre-computed neighbors from the mesh, passed in via the `surface_mesh_neighbors` tensor.

* **Input:** The Trunk Net, implemented as the `NNBasisFunctions` class, is fed a rich set of features for a single point from this stencil. As seen in the `DoMINO` constructor, different networks (`self.nn_basis_surf` and `self.nn_basis_vol`) are created for surface and volume predictions, each tailored to its specific set of input features (e.g., surface predictions can include normals and element areas).

* **Output (Latent Vector):** The `NNBasisFunctions.forward` method processes these inputs through several fully connected layers (`fc1`, `fc2`, `fc3`) and computes a **latent vector**. This output, stored as `basis_f` in the calculation methods, acts as a set of learned coefficients for the basis functions.


---
### Combining for Prediction

The power of the DeepONet structure comes from combining the outputs of both networks.

1.  The feature vector from the **Branch Net** (describing the local shape, `encoding_g`), the latent vector from the **Trunk Net** (describing the query point, `basis_f`), and other encodings like the positional encoding (`encoding_node`) are concatenated into one large feature vector.

2.  This combined vector is then passed through a final set of fully connected layers.

3.  These final layers produce the **solution vector** for that point, containing the predicted values for all physical quantities, such as pressure, wall shear stress, or velocity components.

* **Code Connection:** This entire combination happens inside the `calculate_solution` and `calculate_solution_with_neighbors` functions. You can see the explicit concatenation: `output = torch.cat((basis_f, encoding_node, encoding_g), dim=-1)`. If global parameters are used, `param_encoding` is also concatenated. This combined `output` tensor is then fed into the final prediction module, `agg_model[f](output)`, which is an instance of the `AggregationModel` class.

---
### Final Prediction via Weighted Averaging

To ensure the final prediction is robust and smooth, the model performs the entire prediction process not just for the central point `i`, but for **every point in its dynamically sampled stencil**, including all its neighbors `j`. This results in multiple solution predictions within a small neighborhood, which are then combined.

* **Code Connection:** The logic for this is clearly implemented in the `calculate_solution` and `calculate_solution_with_neighbors` functions (specifically in the `two-loop` or `one-loop` modes).
    * The distance to each neighbor is calculated: `dist = torch.norm(noise, dim=-1, keepdim=True)`.
    * The prediction for each neighbor is weighted by the inverse of this distance: `output_neighbor += agg_model[f](output) * (1.0 / dist)`.
    * Finally, the predictions from the center point and the weighted neighbors are averaged to produce the final result: `output_res = 0.5 * output_center + 0.5 * output_neighbor / dist_sum`.

This inverse distance weighting scheme is a physically intuitive way to ensure that closer neighbors have a stronger influence on the final prediction, effectively smoothing the result.

### Global Geometry Representation:
The Global Geometry Representation refers to the overall shape and structure of the entire object or domain that you are modeling. This representation captures all the geometric details across the entire computational domain.\
Step-by-Step Explanation of Global Geometry Representation:

- **Step 1**: Construct Bounding Boxes
	- A tight-fitting surface bounding box is created around the STL (3D geometry) to hold the geometry.
    - A computational domain bounding box is also defined, which is larger than the surface bounding box to encompass the whole computational domain.
    - Both bounding boxex can be specified in  ```conf.yaml```
- **Step 2**: Project STL Vertices onto Structured Grid
	- The geometric features of the point cloud, such as spatial coordinates, are projected onto an N-dimensional structured grid of resolution m×m×m×f, which is overlaid on the surface bounding box using **learnable point convolution kernels**.
	- The learnable point convolution kernels are created using **differentiable ball query layers**. This means that the method:
    	- Uses a "ball" (a sphere in 3D space) around each point to query or find its neighbors.
    	- The ball query layer is "differentiable," meaning it can be included in the neural network and updated via backpropagation (i.e., during training, the network can learn how to adjust the kernels to improve performance).
    	- The radius of the ball (radius of influence) defines how far around each point we look for neighboring points to include in the convolution. This defines, in fact, how far the geometry can affect the grid. A range of point convolutional kernel sizes can be learned by specifying several radii. Moreover, different kernels are learned to represent information on the surface bounding box and computational domain bounding box. This enables multi-scale learning of geometry encoding by representing both short- and long-range interactions of the surface and flow fields.  The radii of influence are defined as **list** in the ```conf.yaml``` file:
          ```yaml
            volume_radii: [0.1, 0.5]
            surface_radii: [0.05]
          ```
\
          These radius are used in the DoMINO model (```physicsnemo/models/domino/model.py```) to compute two **BQWarp** accordingly:
          

     
```python
        class GeometryRep(nn.Module):
            """Geometry representation from STLs block"""

            def __init__(self, input_features, model_parameters=None):
                super().__init__()
                geometry_rep = model_parameters.geometry_rep

                self.bq_warp = nn.ModuleList()
                self.geo_processors = nn.ModuleList()
                for j, p in enumerate(radii):
                    self.bq_warp.append(
                        BQWarp(
                            input_features=input_features,
                            grid_resolution=model_parameters.interp_res,
                            radius=radii[j],
                        )
                    )
                    self.geo_processors.append(
                        GeoProcessor(
                            input_filters=geometry_rep.geo_conv.base_neurons_out,
                            model_parameters=geometry_rep.geo_processor,
                        )
                    )
```
       
        
- **Step 3**: Use Multi-Resolution Approach for Detailed and Coarse Features
	- The grid resolution in the bounding box determines the level of detail of the geometry: 
    	- Finer resolution captures more detailed features of the geometry.
    	- Coarser resolution captures larger, broader features.
	- A multi-resolution approach is adopted, meaning multiple grids at different resolutions (levels) are maintained to capture both fine and coarse features of the geometry. The number of resolution levels is a parameter that can be adjusted in conf.yaml file as
      ```yaml
      GRID_RESOLUTION = [128, 64, 48]  # Resolution of the interpolation grid 
      ```
  - Currently, the DoMINO model allows specification of a single resolution but this configuration will be provided in a future release.

- **Step 4**: Propagate Geometry Features into the Computational Domain
  - The computational domain is much larger than the surface bounding box, so the geometry information needs to be extended.
	- Geometry features are propagated into the computational domain using two methods: 
    	- As explained in **step 2** Multi-scale **point convolution kernels** project the **geometry information** onto the surface bounding box (**see below the letf figure**).
    	- **Features** from the surface grid of the bounding box (i.e., Gs) are propagated into the computational domain grid (i.e., Gc) using **CNN blocks** that contain convolution, pooling, and unpooling layers (**see below the letf figure**).
          As you see in the cvode snipet above first `BQWarp` is calculated and then  CNN blocks using `GeoProcessor`:
          ```python

          ```
	- The **CNN blocks are iterated** for a specified number of steps to refine the geometry representation. Currently, the DoMINO model is configured to run a single iteration. An option to change this will be provided in ```conf.yaml``` in a future release.

- **Step 5**: Calculate Signed Distance Function (SDF) and its Gradients
	- Additionally, the Signed Distance Function (SDF) and its gradient components are calculated on the computational domain grid.
	- These SDF and gradient values are added to the learned features, providing additional information about the topology of the geometry (i.e., the geometry's shape, distances to surfaces, etc.).

- **Step 6**: Final Global Geometry Representation
	- The final geometry representation of the STL is formed by combining the learned features from the structured grids at different resolutions in both the bounding box and the computational domain.

Once the computational domain is created for each resolution, the next step would be local geometry representation. 

### Local geometry representation
The Local Geometry Representation focuses on the geometry in the immediate vicinity of a sampled point p (the points in simulation mesh). The idea is to understand how the geometry behaves around a specific point and its neighbors, which can be important for accurate predictions. While the Global Geometry Representation gives the big picture, the Local Geometry Representation zooms in on a small region of interest around each sampled point. The key difference is that local geometry represents a smaller, more detailed portion of the global geometry, typically focusing on the small-scale features close to a point. For each sampled point p, neighboring points are sampled randomly around them to form a computational stencil of points similar to finite volume and element methods. The local geometry representation is learned by drawing a subregion around the computational stencil of
p + 1 points. The size of the subregion are defined as **list** in the ```conf.yaml``` file:

  ```yaml
    geometry_local.volume_radii: [0.05, 0.1]
    geometry_local.surface_radii: [0.05]
  ```
Similar to Global geometry representation a point convolution kernel is used here to extract the local features in the subregion from the global geometry representation on the computational domain.
The **BQWarp** are computed for Local geometry representation in ```physicsnemo/models/domino/model.py``` in ```class DoMINO(nn.Module): ```:

**How Does the Multi-Resolution Global Geometry Affect the Local Geometry Representation?**
 - Coarse resolution: At the coarse resolution, you get a broad view of the object. This can give information about the general shape and large-scale features of the geometry (e.g., the overall shape of the object, major boundaries, etc.). When local geometry is extracted from the coarse resolution, the features are relatively less detailed, and it might capture larger, more general features of the object.
 - Fine resolution: At the fine resolution, you get a detailed view of the geometry, capturing small features such as intricate surface details, small holes, or sharp edges. The local geometry representation derived from the fine resolution will be more detailed and capture smaller variations in the geometry near each sampled point.

**Thus, the global multi-resolution geometry allows the local geometry to be learned at different levels of detail, depending on the resolution of the grid that is used to represent the geometry.** 

### Basis Function Neural Network (Latent Vector):
- Once the local geometry representation is built, it is passed through a Basis Function Neural Network.\
  What happens here:    
- The input features (coordinates, SDF, normal vectors, etc. and their fourier features) for each point in the stencil are fed into the Basis Function Neural Network. This is a fully connected neural network that processes these features.
- The network then computes a latent vector for each point in the stencil. A latent vector is a compressed mathematical representation that encodes the important information about each point’s geometry and position.
- Purpose: The latent vector captures the essential characteristics of each point’s geometry in a compact form, which will be used in later steps for predicting the solution at that point.

### Concatenating the Latent Vector with the Local Geometry Encoding:
- After calculating the latent vector for each point, this vector is concatenated with the local geometry encoding — which includes the previously computed information from the surrounding points and the global geometry.
- Why this is done: Concatenating these two representations allows the network to use both the specific local features of each point and the broader context of the surrounding geometry to make predictions.

### Passing Through Additional Neural Network Layers (Solution Prediction):
- The combined information (latent vector + local geometry encoding) is passed through another set of fully connected layers (a new neural network).
- What happens here: These layers process the combined information and predict a solution vector for each point in the stencil. The solution vector could represent various physical quantities such as temperature, pressure, or other simulation results at the sampled point.
- Purpose: This step produces the predicted solution at each point, based on the local and global geometry.

### Aggregation Network:
Aggregation network is a fully connected neural network with a **DeepONet** like structure, **Local geometry rep is branch net and basis functions are trunk net**, which is used to compute solutions.


- Local Geometry Representation is the Branch Net: The branch network in DeepONet is responsible for processing the local geometry representation (like the shape or physical location of the point and its neighbors in the domain). In this case, the branch net processes the features around the sampled point (i) and its neighbors (j).
- Basis Functions are the Trunk Net: The trunk network in DeepONet processes additional data (like the global features or functions) to help represent the solution space better. Here, the basis functions represent mathematical components that help capture the underlying solution in the computational domain.

- Aggregation network computes the solution field on the sampled point, i and its neighbors j. The solutions are then averaged using an inverse distance weighted interpolation scheme.\
In simpler terms:

- The aggregation network computes the solution values at the sampled point (i) and its surrounding neighbors (j). The solution is a value that corresponds to a field (e.g., temperature, pressure) at these points.
- After computing the solution at the sampled point and its neighbors, the results are combined (averaged) using an inverse distance weighting scheme. This means that points closer to the sampled point (i) contribute more to the final solution than points farther away. The "inverse distance" part means that the influence of a neighbor's solution decreases the farther it is from the sampled point.

````python
def mse_loss_fn(output, target, padded_value=-10):
    """
    Compute masked MSE loss, ignoring padded values.
    
    Args:
        output (torch.Tensor): Model predictions
        target (torch.Tensor): Ground truth values
        padded_value (float): Value used for padding (default: -10)
    
    Returns:
        torch.Tensor: Mean squared error loss
    """
    # Move target to same device as output
    target = target.to(output.device)
    # Create mask for non-padded values
    mask = torch.abs(target - padded_value) > 1e-3
    # Compute masked loss
    masked_loss = torch.sum(((output - target) ** 2) * mask) / torch.sum(mask)
    return masked_loss.mean()

def create_dataset(phase):
    """
    Create DoMINO dataset for specified phase (train/val).
    
    Args:
        phase (str): Dataset phase ('train' or 'val')
    
    Returns:
        DoMINODataPipe: Configured dataset
    """
    return DoMINODataPipe(
        DATA_PATHS[phase],
        phase=phase,
        grid_resolution=GRID_RESOLUTION,
        surface_variables=SURFACE_VARS,
        normalize_coordinates=True,
        sampling=True,
        sample_in_bbox=True,
        volume_points_sample=8192,
        surface_points_sample=SURFACE_POINTS_SAMPLE,
        geom_points_sample=60000,
        positional_encoding=False,
        surface_factors=np.load(SURF_SAVE_PATH),
        scaling_type=NORMALIZATION,
        model_type=MODEL_TYPE,
        bounding_box_dims=BOUNDING_BOX,
        bounding_box_dims_surf=BOUNDING_BOX_SURF,
        num_surface_neighbors=NUM_SURFACE_NEIGHBORS,
        gpu_preprocessing=False
        )



def create_dataloaders(rank, world_size):
    """
    Create train and validation dataloaders with distributed sampling.
    
    Args:
        rank (int): Process rank
        world_size (int): Total number of processes
    
    Returns:
        tuple: (train_loader, val_loader, train_sampler, val_sampler)
    """
    # Create datasets
    train_dataset, val_dataset = create_dataset("train"), create_dataset("val")
     
    # Configure distributed samplers if needed
    train_sampler = DistributedSampler(train_dataset, num_replicas=world_size, rank=rank) if world_size > 1 else None
    val_sampler = DistributedSampler(val_dataset, num_replicas=world_size, rank=rank) if world_size > 1 else None
    
    # Create dataloaders
    return (
        DataLoader(train_dataset, batch_size=BATCH_SIZE, sampler=train_sampler, shuffle=train_sampler is None),
        DataLoader(val_dataset, batch_size=BATCH_SIZE, sampler=val_sampler, shuffle=False),
        train_sampler, val_sampler
    )

def create_model(device, rank, world_size):
    """
    Create and configure DoMINO model with distributed training support.
    
    Args:
        device (torch.device): Computation device
        rank (int): Process rank
        world_size (int): Total number of processes
    
    Returns:
        DoMINO: Configured model (wrapped in DistributedDataParallel if world_size > 1)
    """

  
    # Initialize model with configuration
    model = DoMINO(
        input_features=3,
        output_features_vol=None,
        output_features_surf=NUM_SURF_VARS,
        model_parameters=SimpleNamespace(
            activation="relu",
            interp_res=GRID_RESOLUTION,
            surface_neighbors=NUM_SURFACE_NEIGHBORS,
            use_surface_normals=True,
            use_surface_area=True,
            encode_parameters=False,
            positional_encoding=False,
            num_neighbors_surface=7,
            num_neighbors_volume=7,
            combine_volume_surface=False,
            integral_loss_scaling_factor=INTEGRAL_LOSS_SCALING,
            normalization=NORMALIZATION,
            use_sdf_in_basis_func=True,
            geometry_encoding_type= GEOMETRY_ENCODING_TYPE, # geometry encoder type, sdf, stl, both
            geometry_rep=GEOMETRY_REP,
            nn_basis_functions=SimpleNamespace(base_layer=512, fourier_features=True, num_modes=5,activation="relu"),
            position_encoder=SimpleNamespace(base_neurons=512,activation="relu",fourier_features=False,num_modes=5),
            parameter_model=SimpleNamespace(base_layer=512,fourier_features=False,num_modes=5,activation="relu"),
            geometry_local=GEOMETRY_LOCAL,
            local_point_conv=SimpleNamespace(activation="relu"),
            aggregation_model=SimpleNamespace(base_layer=512,activation="relu"),
            model_type=MODEL_TYPE
        ),
    ).to(device)
    
    # Wrap model for distributed training if needed
    if world_size > 1:
        model = DistributedDataParallel(
            model, 
            device_ids=[rank], 
            output_device=rank, 
            find_unused_parameters=True
        )
    
    return model

def run_epoch(train_loader, val_loader, model, optimizer, scaler, device, epoch, best_vloss, rank, world_size):
    """
    Run one training epoch with validation.
    
    Args:
        train_loader (DataLoader): Training data loader
        val_loader (DataLoader): Validation data loader
        model (DoMINO): Model to train
        optimizer (torch.optim.Optimizer): Optimizer
        scaler (GradScaler): Gradient scaler for mixed precision
        device (torch.device): Computation device
        epoch (int): Current epoch number
        best_vloss (float): Best validation loss so far
        rank (int): Process rank
        world_size (int): Total number of processes
    
    Returns:
        float: Validation loss for this epoch
    """
    # Training phase
    model.train()
    train_loss = 0.0
    pbar = tqdm(train_loader, desc=f"Epoch {epoch+1}/{NUM_EPOCHS}") if rank == 0 else train_loader
    
    for batch in pbar:
        # Move batch to device
        batch = dict_to_device(batch, device)
        
        # Forward pass with mixed precision
        with autocast():
            _, pred_surf = model(batch)
            loss = mse_loss_fn(pred_surf, batch["surface_fields"])
        
        # Backward pass with gradient scaling
        scaler.scale(loss).backward()
        scaler.step(optimizer)
        scaler.update()
        optimizer.zero_grad()
        
        # Update loss tracking
        train_loss += loss.detach().item()
        if rank == 0:
            pbar.set_postfix({
                "train_loss": f"{train_loss/(pbar.n+1):.5e}", 
                "lr": f"{optimizer.param_groups[0]['lr']:.2e}",
                "batch_size": batch["surface_fields"].shape[0]
            })
    
    # Compute average training loss
    avg_train_loss = train_loss / len(train_loader)
    
    # Validation phase
    model.eval()
    with torch.no_grad():
        val_loss = sum(
            mse_loss_fn(model(dict_to_device(batch, device))[1], batch["surface_fields"].to(device)).item() 
            for batch in val_loader
        ) / len(val_loader)
    
    # Handle distributed training metrics
    if world_size > 1:
        avg_train_loss, val_loss = [torch.tensor(v, device=device) for v in [avg_train_loss, val_loss]]
        torch.distributed.all_reduce(avg_train_loss, op=torch.distributed.ReduceOp.SUM)
        torch.distributed.all_reduce(val_loss, op=torch.distributed.ReduceOp.SUM)
        avg_train_loss, val_loss = avg_train_loss.item() / world_size, val_loss.item() / world_size
    
    # Save checkpoints on main process
    checkpoint_dir=os.path.join(MODEL_SAVE_DIR, "best_model")
    if rank == 0:
        if val_loss < best_vloss:
            print(f"Checkpoint at Epoch {epoch+1} is stored in {checkpoint_dir}")
            save_checkpoint(
                os.path.join(MODEL_SAVE_DIR, "best_model"), 
                models=model,
                optimizer=optimizer,
                scaler=scaler
            )

        if (epoch + 1) % CHECKPOINT_INTERVAL == 0:
            save_checkpoint(
                MODEL_SAVE_DIR, 
                models=model, 
                optimizer=optimizer, 
                scaler=scaler, 
                epoch=epoch
            )
    
    return val_loss

def train(model, device, rank, world_size):
    """
    Function that orchestrates the training process.
    Handles distributed training setup, model creation and training loop.
    """


    # Create output directory on main process
    os.makedirs(MODEL_SAVE_DIR, exist_ok=True) if rank == 0 else None
    
    # Set up data
    train_loader, val_loader, train_sampler, val_sampler = create_dataloaders(rank, world_size)

    # Grab a single batch from the training loader for inspection
    sample = next(iter(train_loader))
    
    print("\n🔍 Sample batch from `train_loader` (returned by `create_dataloaders`):")
    print("----------------------------------------------------------")
    print(f"Batch contains {len(sample)} entries (features/fields).")
    print("Each key corresponds to a data field in the batch:\n")
    
    for key, value in sample.items():
        dtype = type(value)
        shape = getattr(value, "shape", None)
        print(f"🧷 Key: '{key}' | Type: {dtype.__name__} | Shape: {shape}")
    
    
    optimizer = apex.optimizers.FusedAdam(model.parameters(), lr=0.001)
    
    # Initialize learning rate scheduler and gradient scaler
    #scheduler = torch.optim.lr_scheduler.MultiStepLR(optimizer, milestones=[1, 2], gamma=0.5)
    scaler = GradScaler()
    
    # Training loop
    best_vloss = float('inf')
    for epoch in range(NUM_EPOCHS):
        if world_size > 1:
            train_sampler.set_epoch(epoch)
        best_vloss = min(
            best_vloss, 
            run_epoch(
                train_loader, val_loader, model, optimizer, 
                scaler, device, epoch, best_vloss, rank, world_size
            )
        )
        #scheduler.step()
````

#### Implementation Overview

The testing is implemented through several key components:

1. **Function test_step**
    Within the **test_step** function, several key operations execute in sequence. Initially, torch.no_grad() is used to disable gradient tracking in PyTorch, optimizing performance by saving memory and computation time as gradients are unnecessary during inference. 
Next, the necessary data is prepared by extracting inputs like air density, stream velocity, geometry coordinates, bounding box grid information (surf_grid), and the Signed Distance Field (SDF) from the data_dict; the SDF is particularly important as it helps the model understand the position of points relative to the geometry surface. 
Following this, a global geometry encoding is generated using model.geo_rep_surface, which takes normalized geometry points, the grid, and the SDF to create a comprehensive representation of the overall shape. 
Surface-specific data, including mesh points, normals, areas, and neighbor details (found via methods like KDTree during preprocessing), are then extracted. 
To refine the focus, model.geo_encoding_local_surface extracts relevant local geometric features from the global encoding specifically for the surface points where predictions are needed. 
Positional awareness is added using model.position_encoder to encode the relative location of surface points. 
The core prediction then occurs via model.calculate_solution_with_neighbors, combining local geometry, positional encoding, surface point details, neighbor information, and flow conditions to estimate the target surface fields like pressure coefficient or wall shear stress. Since the model output is normalized, a final un-normalization step converts these predictions back into physical units using the provided surf_factors, stream velocity, and air density. The function concludes by returning the predicted surface fields (prediction_surf), as this specific code path concentrates only on surface predictions.

````python
from utils import *

def test_step(model, data_dict, surf_factors, device):
    """
    Runs a single, clean test step on the model.
    """
    with torch.no_grad():
        # 1. Prepare data for model
        data = dict_to_device(data_dict, device)
        
        # **FIX:** Ensure tensor inputs are contiguous in memory before the model call.
        geo_centers_norm = normalize_to_box(
            data["geometry_coordinates"], 
            data["surface_min_max"][:, 0], 
            data["surface_min_max"][:, 1]
        ).contiguous() # Add .contiguous() here

        surf_grid_contiguous = data["surf_grid"].contiguous()
        sdf_surf_grid_contiguous = data["sdf_surf_grid"].contiguous()

        # 2. Run model inference with the contiguous tensors
        encoding_g_surf = model.geo_rep_surface(
            geo_centers_norm, 
            surf_grid_contiguous, 
            sdf_surf_grid_contiguous
        )
        
        # The rest of the function remains the same...
        geo_encoding_local = model.geo_encoding_local(0.5 * encoding_g_surf, data["surface_mesh_centers"].contiguous(), surf_grid_contiguous, mode="surface")
        pos_encoding = model.position_encoder(data["pos_surface_center_of_mass"].contiguous(), eval_mode="surface")
        
        prediction_tensor = model.calculate_solution_with_neighbors(
            data["surface_mesh_centers"].contiguous(), geo_encoding_local, pos_encoding,
            data["surface_mesh_neighbors"].contiguous(), data["surface_normals"].contiguous(), data["surface_neighbors_normals"].contiguous(),
            torch.unsqueeze(data["surface_areas"], -1), torch.unsqueeze(data["surface_neighbors_areas"], -1),
            data["stream_velocity"], data["air_density"]
        )

        # 3. Post-process the output
        prediction_surf = unscale_surface_prediction(
            prediction_tensor, surf_factors,
            data["stream_velocity"][0, 0].cpu().numpy(),
            data["air_density"][0, 0].cpu().numpy()
        )
        
    return None, prediction_surf # prediction_vol is not calculated

def test(filepath, dirname, model):
    """
    High-level pipeline to manage the entire testing process for a single case.
    """
    # 1. Setup & Configuration
    # Assumes constants like SURFACE_VARS, CHECKPOINT_DIR, etc., are loaded from a config file
    surf_factors = np.load(os.path.join("outputs", PROJECT_NAME, "surface_scaling_factors.npy"))
    best_checkpoint = torch.load(best_model_path)
    model.load_state_dict(best_checkpoint)  # Load the model state
    
    #model = load_model(model, best_model_path)
    tag, stl_path = get_test_case_info(filepath)
    vtp_pred_save_path = SAVE_PATH / f"boundary_{tag}_predicted.vtp"
    
    # 2. Data Loading & Pre-processing
    stream_velocity = 30
    stl_data = process_stl_file(stl_path)
    sdf_data = create_sdf_grid(stl_data["vertices"], stl_data["faces_indices"], GRID_RESOLUTION, BOUNDING_BOX_SURF)
    vtp_data = process_vtp_data(filepath, NUM_SURFACE_NEIGHBORS, SURFACE_VARS, BOUNDING_BOX_SURF)
    data_dict = assemble_data_dict(stl_data, sdf_data, vtp_data, stream_velocity, AIR_DENSITY)

    # 3. Run Inference
    _, prediction_surf = test_step(model, data_dict, surf_factors, device)
    
    # 4. Post-process, Evaluate, and Save Results
    force_x_pred, force_x_true = calculate_forces(
        prediction_surf, vtp_data["fields"], vtp_data["normals"], vtp_data["sizes"]
    )
    print(f"Directory: {dirname}, Predicted Force: {force_x_pred:.4f}, True Force: {force_x_true:.4f}")
    
    save_predictions_to_vtp(
        vtp_data["polydata"], prediction_surf, SURFACE_VARS, vtp_pred_save_path
    )
````
